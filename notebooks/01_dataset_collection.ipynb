```python
mozilla_std = pd.DataFrame({
    'bug_id':mozilla_df['bug_id'],
    'title':mozilla_df['short_description'],
    'description':mozilla_df['long_description'],
    'severity':mozilla_df['severity_category'],
    'resolution':mozilla_df['resolution_category'],
    'status':mozilla_df['status_category'],
    'source_dataset':'Mozilla'
}
    
)

eclipse_std = pd.DataFrame({
    'bug_id':eclipse_df['bug_id'],
    'title':eclipse_df['short_description'],
    'description':eclipse_df['long_description'],
    'severity':eclipse_df ['severity_category'],
    'resolution':eclipse_df['resolution_category'],
    'status':eclipse_df['status_category'],
    'source_dataset':'Eclipse'
}
    
)

apache_std = pd.DataFrame({
    'bug_id':apache_df['key'],
    'title':apache_df['summary'],
    'description':apache_df['description'],
    'severity':apache_df ['priority.name'],
    'resolution':apache_df['resolution.name'],
    'status':apache_df['statusCategory.name'],
    'source_dataset':'Apache'
    
})

print('Mozilla:',mozilla_std.shape)
print('Eclipse:',eclipse_std.shape)
print('Apache:',apache_std.shape)
```

```python
mozilla_std['stack_trace'] = None
eclipse_std['stack_trace'] = None
apache_std['stack_trace'] = None

mozilla_std['product_component'] = mozilla_df['product_name']
eclipse_std['product_component'] = eclipse_df['product_name']
apache_std['product_component'] = apache_df['project.name']


print(mozilla_std.shape)
print(eclipse_std.shape)
print(apache_std.shape)

mozilla_std.head()
```

```python
knowledge_base_df = pd.concat([mozilla_std,eclipse_std,apache_std], ignore_index = True)
print("Comobo Size " , knowledge_base_df.shape)
knowledge_base_df['source_dataset'].value_counts()
```
