# TASK 7:DATA CLEANING

```python
print(knowledge_base_df.isnull().sum())
print("\n\nDuplicate id : ",knowledge_base_df['bug_id'].duplicated().sum())
print("\n\nShape before cleaning : ",knowledge_base_df.shape)

```

```python
knowledge_base_clean = knowledge_base_df.dropna(subset = ['title','description'])

knowledge_base_clean = knowledge_base_clean.fillna({
    'severity':'Unknown',
    'resolution':'Not Specificed',
    'status':'Unknown'
})

print("Shape after cleaning :",knowledge_base_clean.shape)
print("\nRemaining nulls : \n",knowledge_base_clean.isnull().sum())
```

```python
import re

def clean_text(text):
    if pd.isna(text):
        return(text)
    text =  str(text)
    text = re.sub(r'<[^>]+>',' ',text)
    text = re.sub(r'\s+',' ',text)
    return text.strip()

knowledge_base_clean['title'] = knowledge_base_clean['title'].apply(clean_text)
knowledge_base_clean['description'] = knowledge_base_clean['description'].apply(clean_text)

print('Cleaning applied . A Sample of that :')
knowledge_base_clean[['title','description']].head(3)
    
```

# TASK 8:Chunking with  Langchain

```python
import sys
!{sys.executable} -m pip install langchain langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=500 , chunk_overlap=50)

knowledge_base_clean['full_text'] = knowledge_base_clean['title'] + ". " + knowledge_base_clean['description']

sample_text = knowledge_base_clean['full_text'].iloc[0]
chunks = splitter.split_text(sample_text)

print("Original Text Length :",len(sample_text))
print("Numberof chunks :",len(chunks))
print("\nFirst Chunk:\n",chunks[0])
```

```python
longer_text = knowledge_base_clean['full_text'].iloc[1]
chunks = splitter.split_text(longer_text)

print("Original Text Length :",len(longer_text))
print("Numberof chunks :",len(chunks))
print("\n--- Chunk 1 ---\n",chunks[0])
print("\n--- Chunk 2 ---\n",chunks[1] if len(chunks)>1 else "(only 1 chunk)")
```

```python
longest_idx = knowledge_base_clean['full_text'].str.len().idxmax()
longest_text=knowledge_base_clean['full_text'].loc[longest_idx]

print("Length of longest text:", len(longest_text))  
chunks = splitter.split_text(longest_text)

print("Length of longest text:",len(chunks))
print("Number of chunks :",len(chunks))
print("\n--- Chunk 1 ---\n",chunks[0])
print("\n--- Chunk 2 ---\n",chunks[1] if len(chunks)>1 else "(only 1 chunk)")
```

```python
# Cap any single text at 3000 characters before chunking (keeps outliers from exploding chunk count)
knowledge_base_clean['full_text'] = knowledge_base_clean['full_text'].str[:3000]
```

```python
all_chunks = []

for idx , row  in knowledge_base_clean.iterrows():
    text_chunks = splitter.split_text(row['full_text'])
    for i , chunk in enumerate(text_chunks):
        all_chunks.append({
            'chunk_id':f"{row['bug_id']}_chunk{i}",
            'bug_id':row['bug_id'],
            'chunk_text':chunk,
            'title':row['title'],
            'severity':row['severity'],
            'resolution':row['resolution'],
            'status':row['status'],
            'source_dataset':row['source_dataset']
        })

chunks_df = pd.DataFrame(all_chunks)
print("Total chunks created:",chunks_df.shape[0])
print("From original rows:", knowledge_base_clean.shape[0])
chunks_df.head(55)
```

```python
import chromadb

client = chromadb.Client()
collection = client.create_collection(name="bug_knowledge_base")

# Convert TF-IDF sparse matrix to lists for ChromaDB
embeddings_list = embeddings.toarray().tolist()

# Add in batches (ChromaDB works better with reasonable batch sizes)
batch_size = 5000
for i in range(0, len(chunks_df), batch_size):
    batch_end = min(i + batch_size, len(chunks_df))
    collection.add(
        ids=chunks_df['chunk_id'].iloc[i:batch_end].tolist(),
        embeddings=embeddings_list[i:batch_end],
        documents=chunks_df['chunk_text'].iloc[i:batch_end].tolist(),
        metadatas=chunks_df[['bug_id', 'title', 'severity', 'resolution', 'status', 'source_dataset']].iloc[i:batch_end].to_dict('records')
    )
    print(f"Added batch {i} to {batch_end}")

print("Total items in collection:", collection.count())
```
